```python
# === Cell: استيراد المكتبات ===
import numpy as np                      # np: عمليات عشوائية ومصفوفات
import pandas as pd                     # pd: جداول بيانات (DataFrame)
import matplotlib.pyplot as plt         # رسم بياني أساسي
import seaborn as sns                   # رسوم إحصائية أجمل
%matplotlib inline
sns.set(style="whitegrid")              # يضبط شكل الرسوم

# === Cell: توليد داتا اصطناعية بسيطة (لتتدرّبي) ===
np.random.seed(42)                      # يضمن نفس الأرقام عند إعادة التشغيل
n = 500                                 # عدد السجلات التجريبي
ages = np.random.randint(18, 90, n)     # أعمار عشوائية بين 18 و 89
gender = np.random.choice(['M','F'], n, p=[0.48,0.52])  # جنس بنسب قريبة للواقع
claim_amount = np.round(np.random.exponential(scale=1200, size=n) + 50, 2)
# ملاحظة: exponential تعطي ذيل طويل مناسب للأموال
fraud_prob = 0.01 + 0.000005*claim_amount   # قاعدة بسيطة للاحتيال (تعليمي)
fraud = np.random.binomial(1, np.clip(fraud_prob,0,0.5))  # 0/1 label
df = pd.DataFrame({
    "age": ages,
    "gender": gender,
    "claim_amount": claim_amount,
    "fraud": fraud
})
df.head()  # يعرض أول 5 صفوف
```

```python
# Basic stats - EDA
print("shape:", df.shape)                # (n_rows, n_cols)
print(df.describe().round(2))            # count, mean, std, min, 25%, 50%, 75%, max

# رياضياً: المتوسط و التباين و الانحراف
mean_claim = df['claim_amount'].mean()   # المتوسط: μ = (1/n) Σ x_i
var_claim = df['claim_amount'].var()     # التباين: s^2 = (1/(n-1)) Σ (x_i-μ)^2
std_claim = df['claim_amount'].std()     # الانحراف: s = sqrt(var)
print("mean:", mean_claim, "var:", var_claim, "std:", std_claim)

# correlation بين age و claim_amount
corr = df[['age','claim_amount']].corr().iloc[0,1]  # معامل الارتباط بيرسون
print("Pearson correlation (age vs claim):", round(corr,3))
```

```python
# Histogram - توزيع مبلغ المطالبة
plt.figure(figsize=(8,4))
sns.histplot(df['claim_amount'], bins=50, kde=True)
plt.title("Distribution of Claim Amount")
plt.xlabel("Claim amount")
plt.show()

# Boxplot - مقارنة مبلغ المطالبة حسب fraud
plt.figure(figsize=(6,4))
sns.boxplot(x='fraud', y='claim_amount', data=df)
plt.title("Claim amount by fraud")
plt.show()

# Scatter - age vs claim_amount colored by fraud
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='claim_amount', hue='fraud', data=df, alpha=0.6)
plt.title("Age vs Claim amount (colored by fraud)")
plt.show()
```

```python
df['claim_log'] = np.log1p(df['claim_amount'])  # يخفف التأثير العددي للقيم الكبيرة
```

```python
import sqlite3

# أنشئ اتصال b sqlite db
conn = sqlite3.connect('claims.db')  # ينشئ file claims.db
df.to_sql('claims', conn, if_exists='replace', index=False)  # يرفع DataFrame كجدول SQL
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# تجهيز X, y (مثال: استخدام "claim_log" إن أنشأناها)
df['claim_log'] = np.log1p(df['claim_amount'])
X = df[['age','claim_log']]   # مثال ميزتين بسيطتين
y = df['fraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2)

model = LogisticRegression(class_weight='balanced', max_iter=1000)  # class_weight يساعد بتوازن الفئات
model.fit(X_train, y_train)   # تدريب النموذج

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:,1]
print(classification_report(y_test, y_pred, digits=4))
print("ROC AUC:", round(roc_auc_score(y_test, y_proba),4))
```
